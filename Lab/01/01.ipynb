# Q1

```python
import random
import pandas as pd
```

```python
testScores, readingSkill, writingSkills, classParticipation, studyHrs = [[random.randint(1,5) for _ in range(10)] for _ in range (5)]
names = ["Ali", "Sara", "Omar", "Ayesha", "Bilal", "Fatima", "Hassan", "Noor", "Usman", "Zara"]
```

```python
print(testScores)
print(readingSkill)
print(writingSkills)
print(classParticipation)
print(studyHrs)
```

```python
df = {'test_scores':testScores, 'reading_skill':readingSkill, 'writing_skill':writingSkills, 'class_participation': classParticipation, 'study_hours': studyHrs}
dataframe = pd.DataFrame(df)
dataframe.insert(0, 'name', names)
dataframe = dataframe.set_index('name')
print("1st DataFrame -> Cause of Examination failure")
print(dataframe)
```

```python
BodyFat = [random.randint(8,20) for _ in range(10)] 
Height = [random.randint(140,190) for _ in range(10)] 
sex = [random.choice(["male", "female"]) for _ in range(10)]
Athleticism = [random.randint(6,10) for _ in range(10)] 
```

```python
df = {'Body fat %': BodyFat, 'Height(cm)':Height, 'Sex':sex, 
      'Athletic Ability': Athleticism}
dataframe = pd.DataFrame(df)
dataframe.insert(0, 'name', names)
dataframe = dataframe.set_index('name')
print("2nd DataFrame -> Athlete Data")
print(dataframe)
```

```python
book_names = [
    "The Midnight Library",
    "Shadows of Eternity",
    "Whispers in the Wind",
    "The Crimson Crown",
    "Echoes of Silence",
    "The Last Horizon",
    "Secrets of the Abyss",
    "The Forgotten Path",
    "Wings of Destiny",
    "The Broken Oath"
]
genre = [random.choice(["fiction", "non-fiction"]) for _ in range(10)]
sales = [random.randint(1, 100) for _ in range(10)]
df = {"Book Title":book_names, "Genre": genre, "Sales(k)":sales}
dataframe = pd.DataFrame(df)
dataframe = dataframe.set_index("Book Title")
print(dataframe)
```
